```python
def List1D(val, n):
    lst = [val] * n
    return lst
```

```python
def List2D(val, n, m):
    lst = [ [val]*m for _ in range(n) ]
    return lst
```

```python
def List3D(val, n, m, l):
    lst = [ [ [val]*l for _ in range(m) ] for p in range(n) ]
    return lst
```

```python
def makeBrain(inputs, outputs, brainWidth, brainLength, startProb):
    
    inp = List1D(0, inputs)
    out = List1D(0, outputs)
    nodes = List2D(0, brainLength, brainWidth)
    nodes.insert(0, inp)
    nodes.append(out)
    
    links = List3D(startProb, brainLength-1, brainWidth, brainWidth)
    inpLinks = List2D(startProb, brainWidth, inputs)
    outLinks = List2D(startProb, outputs, brainWidth)
    links.insert(0, inpLinks)
    links.append(outLinks)
    
    return nodes, links
```

```python
def updateBrain(nodes, links, recents=None):
    
    if (not recents):
        recents = []
    
    recents.insert(0, [])
    
    newNodes = copy.deepcopy(nodes)
    
    changed = 0
    
    for i in range(len(nodes)-1):
        for j in range(len(nodes[i+1])):
            for k in range(len(links[i][j])):
                n = random.randint(1, 100000)
                if (nodes[i][k] == 1 and (n/100000) <= links[i][j][k]):
                    newNodes[i+1][j] = 1
                    recents[0].append([i, j, k])
                    changed = 1
            
            if (changed == 0):
                newNodes[i+1][j] = 0
            changed = 0
                    
    return newNodes, recents
```

```python
def modBrain(nodes, links, recents, good, strength, dropoff):
    
    newRecents = copy.deepcopy(recents)
    
    strength = 1-strength
    dropoff = 1-dropoff
    
    brainLength = len(nodes)-2
    
    numRecents = len(recents)
    
    if (good == True):
        for i in range(numRecents):
            for j in range(len(recents[i])):
                item = recents[i][j].copy()
                if (item[0] >= (brainLength-i)):
                    val = 1 - links[item[0]][item[1]][item[2]]
                    val = val*strength
                    val = 1 - val
                    links[item[0]][item[1]][item[2]] = val
                    newRecents[i].remove(item)
            strength = 1 - (1 - strength)*dropoff
    else:
        for i in range(numRecents):
            for j in range(len(recents[i])):
                item = recents[i][j].copy()
                if (item[0] >= (brainLength-i)):
                    val = links[item[0]][item[1]][item[2]]
                    val = val*strength
                    links[item[0]][item[1]][item[2]] = val
                    newRecents[i].remove(item)
            strength = 1 - (1 - strength)*dropoff
    
    return links, newRecents
```
